```python
# Fill Postal Code missing values with mode (common choice)

df['Postal Code'].fillna(df['Postal Code'].mode()[0], inplace=True)
```

```python
# Drop rows with missing critical fields (if any remain)

df.dropna(inplace=True)
```

```python
# Remove duplicate rows

df.drop_duplicates(inplace=True)
print("After removing duplicates:", df.shape)
```

```python
# Function to remove outliers

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]
```

```python
# Apply to numerical columns

for col in ['Sales', 'Profit', 'Discount']:
    df = remove_outliers_iqr(df, col)

print("After outlier removal:", df.shape)
```

```python
# Mean, Median, Std

print("Sales Mean:", df['Sales'].mean())
print("Sales Median:", df['Sales'].median())
print("Sales Std:", df['Sales'].std())


```

```python
# Total Sales by Region
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print("Sales by Region:\n", region_sales)

# Total Sales by Category
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print("\nSales by Category:\n", category_sales)
```

```python
# Sales by Region
sns.barplot(x=region_sales.index, y=region_sales.values, palette="Blues_d")
plt.title("Total Sales by Region")
plt.xticks(rotation=45)
plt.ylabel("Sales")
plt.show()
```

```python
# Sales by Category
sns.barplot(x=category_sales.index, y=category_sales.values, palette="Greens_d")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```
